```python
X = fetal_data.drop(["fetal_health"] , axis=1 )
Y = fetal_data["fetal_health"]
X.shape
```

```python
Y.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_columns = X.columns
sds = StandardScaler(copy=True , with_mean=True , with_std=True)
X = sds.fit_transform(X)
X = pd.DataFrame(X , columns=X_columns)
X
```

```python
x_train , x_test , y_train , y_test = train_test_split(X , Y ,test_size=0.2 , random_state=42)
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
pipeline_LR = Pipeline([("LR_classifier" , LogisticRegression(random_state=11))])
pipeline_DT = Pipeline([("DT_classifier" , DecisionTreeClassifier(random_state=11))])
pipeline_RF = Pipeline([("RF_classifier" , RandomForestClassifier())])
pipeline_svc = Pipeline([("SV_classifier" , SVC())])
pipelines = [pipeline_LR , pipeline_DT , pipeline_RF , pipeline_svc]
pipe_dict = {0 : "LogisticRegression" , 1 : "DecisionTreeClassifier" , 2 : "RandomForestClassifier" , 3 : "SVC"}
```

```python
for pipe in pipelines:
    pipe.fit(x_train , y_train)
```

```python
print(pipeline_RF.score(x_train , y_train))
print(pipeline_RF.score(x_test , y_test))
```

```python
for i in pipelines:
    print(f"the score of {i} : ",i.score(x_train , y_train))
    print(i.score(x_test , y_test))
```

```python
pred_rf = pipeline_RF.predict(x_test)
pred_rf
```

```python
from sklearn.metrics import precision_score , recall_score , confusion_matrix , classification_report , accuracy_score
```

```python
print(classification_report(y_test , pred_rf))
```

```python
con_matrix = confusion_matrix(y_test , pred_rf)
con_matrix
```

```python
plt.figure(figsize=(10 , 6))
plt.style.use("classic")
sns.heatmap(con_matrix / np.sum(con_matrix)  , annot=True)
plt.show()
```
